## Download 100 copies of 10-K from Edgar

```python
import requests
import time  
import random
```

```python
# read the file company.idx

indexFile = open('C:/Users/jlxle/Desktop/company.idx').readlines()  
print(indexFile[:12])
```

```python
# Download the filtered files  

count = 0
for i in indexFile[10:6000]:
    if "NT 10-K" in i:
        continue
    if "10-K/A" in i:  # exclude NT 10-K, 10-K/A
        continue
    if '10-K' in i:
        i1 = i.split()   
        i2 = i1[-1].split('/')[-1]
        r = requests.get('https://www.sec.gov/Archives/'+ i1[-1])
        f_name = i1[-4] + '--' + i1[-3] + '--' + i1[-2] + '--' + i2.replace('.txt','.html')  # save as .html
        with open('C:/Users/jlxle/Desktop/files/'+ f_name ,"w") as f: 
            f.write(r.text)              
        f.close()        
        count += 1      
        time.sleep(random.random()*3)
        print(count)
        print(f_name)
    if count == 100:                     
        break
```
